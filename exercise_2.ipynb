```python
import math

print("press 1 for rectangle and press 2 for triangular")
choice=0

#כל עוד לא בחרו 3 המערכת תמשיך להציע למשתמש אפשרויות
while choice!=3:
    choice = int(input("choose option 1 or option 2 to exit press 3"))
    if choice==1:
        height=int(input("enter tower height"))
        width=int(input("enter tower width"))
        if height==width:
            print(height*width)
        elif abs(height-width)>5:
            print(height*width)
        else:
            print(2*(height+width))
    if choice==2:
        height=int(input("enter tower height"))
        width=int(input("enter tower width"))
        user_input = input("If you want triangle scope Enter 'yes' otherwise enter 'no'").lower() 
        if user_input=='yes':
            #משפט פיתגורס לחישוב היקף משולש 
            equal_side_length = math.sqrt((width / 2) ** 2 + height ** 2)
            perimeter = width + 2 * equal_side_length
            print("the scope of the triangle is:", perimeter)
            
        else:
            user_input = input("If you want triangle area Enter 'yes' otherwise enter 'no'").lower()
            if user_input=='yes':
                if width%2==0 or width > height *2:
                    print("can't print the triangle")
                elif width%2!=0 and width<2*height:
                    counter=0
                    # מערך לשמירת מספר שורות של כוכביות בכל רמה
                    arr=[]
                    # בדיקה כמה מספרים אי-זוגיים קיימים בגובה המגדל 
                    for i in range(2,width):
                        if i%2!=0:
                            arr.append(i)
                            counter+=1
                            
                    # שמירת גובה המגדל פחות השורה הראשונה והאחרונה        
                    new_height=height-2
                    #כמה שורות בכל קבוצה במגדל   
                    divide=int(new_height/counter)
                    #כמה שורות בקבוצה העליונה של המגדל  
                    counterTop=divide
                    tmp=divide

                    #בדיקה האם צריך עוד שורות בקבוצה העליונה
                    while tmp*counter!=new_height:
                        counter+=1
                        counterTop+=1
                        tmp = new_height / counter

                    spaces=int(width/2)

                    #הדפסת שורה ראשונה
                    print(" "*spaces,end="")
                    print("*")
                    spaces-=1

                    #הדפסת שורות ביניים של המגדל
                    for i in range(len(arr)):
                        if i==0:
                            for j in range(counterTop):
                                print(" "*spaces,end="")
                                for t in range(arr[i]):
                                     print("*",end="")
                                print(" ")
                            spaces-=1
                        else:
                            for j in range(divide):
                                print(" "*spaces,end="")
                                for t in range(arr[i]):
                                     print("*",end="")
                                print(" ")
                            spaces-=1
                    #הדפסת שורה אחרונה
                    print("*"*width,end="")
```
